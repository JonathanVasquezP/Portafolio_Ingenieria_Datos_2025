```python
# Importar librería necesaria
from google.colab import files
import pandas as pd

# Abrir ventana para subir archivo CSV
uploaded = files.upload()

# uploaded es un diccionario {nombre_archivo: contenido}
# Obtenemos el nombre del archivo subido
archivo = list(uploaded.keys())[0]

# Cargar el CSV en un DataFrame de pandas
df2 = pd.read_csv(archivo, sep=';')

# Mostrar las primeras filas
df2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("whitegrid")

plt.figure(figsize=(8,5))
sns.scatterplot(data=df2, x='Variable_X', y='Variable_Y')
plt.title("Dispersión df2: Variable_X vs Variable_Y")
plt.xlabel("Variable_X")
plt.ylabel("Variable_Y")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.regplot(data=df1, x='Horas_Estudio', y='Nota_Final', ci=None, line_kws={"color":"red"})
plt.title("Horas de Estudio vs Nota Final")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(data=df2, x='Variable_X', y='Variable_Y', ci=None, line_kws={"color":"red"})
plt.title("Variable_X vs Variable_Y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def verificar_linealidad(df, x_col, y_col):
    X = df[[x_col]]
    y = df[y_col]

    modelo = LinearRegression()
    modelo.fit(X, y)
    y_pred = modelo.predict(X)

    mse = mean_squared_error(y, y_pred)
    print(f"MSE para {x_col} vs {y_col}: {mse:.3f}")

    if mse < (0.1 * y.var()):
        print("➡️ La relación parece lineal. Modelo sugerido: Regresión Lineal.\n")
    else:
        print("➡️ La relación parece no lineal. Modelo sugerido: Polinómico o Árbol.\n")
```

```python
verificar_linealidad(df1, 'Horas_Estudio', 'Nota_Final')
verificar_linealidad(df2, 'Variable_X', 'Variable_Y')
```
