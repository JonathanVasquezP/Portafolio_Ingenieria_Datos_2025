```python
# Clase Auto
class Auto:
    def __init__(self, color, peso, tamaño, alto, largo, ruedas, puertas, tipo):
        self.color = color
        self.peso = peso
        self.tamaño = tamaño
        self.alto = alto
        self.largo = largo
        self.ruedas = ruedas
        self.puertas = puertas
        self.tipo = tipo
        self.estado = "detenido"  # Estado inicial

    # Métodos
    def arrancar(self):
        if self.estado == "detenido":
            self.estado = "circulando"
            print("🚗 El auto ha arrancado y está circulando.")
        else:
            print(f"⚠️ No se puede arrancar porque el auto está {self.estado}.")

    def frenar(self):
        if self.estado == "circulando":
            self.estado = "detenido"
            print("🛑 El auto ha frenado y está detenido.")
        else:
            print(f"⚠️ No se puede frenar porque el auto está {self.estado}.")

    def acelerar(self):
        if self.estado == "circulando":
            print("🏎️ El auto está acelerando.")
        else:
            print(f"⚠️ No se puede acelerar porque el auto está {self.estado}.")

    def girar(self, direccion):
        if self.estado == "circulando":
            print(f"↪️ El auto está girando a la {direccion}.")
        else:
            print(f"⚠️ No se puede girar porque el auto está {self.estado}.")

    def estacionar(self):
        if self.estado == "circulando":
            self.estado = "estacionado"
            print("🅿️ El auto está estacionado.")
        else:
            print(f"⚠️ No se puede estacionar porque el auto está {self.estado}.")

    def dañar(self):
        self.estado = "dañado"
        print("💥 El auto está dañado.")

    def reparar(self):
        if self.estado == "dañado":
            self.estado = "detenido"
            print("🔧 El auto ha sido reparado y está detenido.")
        else:
            print("✅ El auto no está dañado.")

    def estado_actual(self):
        print(f"\n🔍 El estado actual del auto es: **{self.estado.upper()}**")


# Pedir datos al usuario
print("🛠️ Configuración de tu Auto\n")
color = input("Color: ")
peso = float(input("Peso (kg): "))
tamaño = input("Tamaño (Pequeño, Mediano, Grande): ")
alto = float(input("Alto (m): "))
largo = float(input("Largo (m): "))
ruedas = int(input("Cantidad de ruedas: "))
puertas = int(input("Cantidad de puertas: "))
tipo = input("Tipo (Sedán, SUV, Pickup, etc.): ")

# Crear la instancia
mi_auto = Auto(color, peso, tamaño, alto, largo, ruedas, puertas, tipo)

# Simulación de acciones
print("\n🔧 Simulación en curso...\n")
mi_auto.arrancar()
mi_auto.acelerar()
mi_auto.girar("derecha")
mi_auto.frenar()
mi_auto.estacionar()

# Mostrar estado final
mi_auto.estado_actual()
```
